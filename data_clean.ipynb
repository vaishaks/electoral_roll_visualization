```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('RptPC_WISE_TURNOUT.csv')
df.head()
```

```python
df.shape
```

```python
state = ''
for x in xrange(df.shape[0]):
    try:
        np.isnan(df['State'][x])
        df['State'][x] = state
    except TypeError:
        state = df['State'][x]
```

```python
df.head()
```

```python
df['PC Id'] = np.nan
```

```python
df.head()
```

```python
ix = [not x for x in df['PC Name'].isin(['State Total: '])]
df1 = df[ix]
```

```python
df1.head()
```

```python
df1.shape
```

```python
df1.head()
```

```python
df2 = df1.reset_index(drop=True)
```

```python
df2.head()
```

```python
df2.tail()
```

```python
df3 = df2[df2['State'] != 'India: ']
df3.tail()
```

```python
for x in xrange(df3.shape[0]):
    df3['PC Id'][x] = df3['PC Name'][x].split(' - ')[0]
    df3['PC Name'][x] = df3['PC Name'][x].split(' - ')[1]
    df3['Male Turnout'][x] = df3['Male Turnout'][x].split()[0]
    df3['Female Turnout'][x] = df3['Female Turnout'][x].split()[0]
    df3['Total Turnout'][x] = df3['Total Turnout'][x].split()[0]
```

```python
df3.head()
```

```python
df3.tail()
```

```python
df3.to_csv('states_pc_mf.csv', index=False)
```
